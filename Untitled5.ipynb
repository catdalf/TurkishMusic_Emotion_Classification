# Data Preprocessing

```python
#Datapreprocessing part
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# Drop the target column from the features (X)
X = df.drop('Class', axis=1)

# Set the target variable (y) to the 'Class' column
y = df['Class']

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2,random_state=42)

```

Since my target column was 'Class' which includes various emotions I have declared it as y and other columns which are explanatory labeled as X and I removed column'Class' from them.

# Model Selection

```python
#Model Selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import xgboost as xgb

# Create and train models
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train, y_train)

qda_model = LinearDiscriminantAnalysis()
qda_model.fit(X_train, y_train)

svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

lr_model = LogisticRegression(penalty='l2', solver='liblinear')
lr_model.fit(X_train, y_train)

xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)
```

# Hyperparameter Optimization

```python
#Hyperparamter optimization
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the pipeline with a different solver and increased max_iter
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(solver='lbfgs', max_iter=1000))
])

param_grid = {'classifier__C': [0.001, 0.01, 0.1, 0.15, 0.2, 0.25, 0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1, 10, 100]}

grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters and accuracy
best_C = grid_search.best_params_['classifier__C']
best_accuracy = grid_search.best_score_
print("Best C:", best_C)
print("Best Accuracy:", best_accuracy)
```

# Model Evaluation

```python
#Model Evaluation
from sklearn.metrics import accuracy_score

# Evaluate models
y_pred_rf = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)

y_pred_qda = qda_model.predict(X_test)
accuracy_qda = accuracy_score(y_test, y_pred_qda)
print("LDA Accuracy:", accuracy_qda)

y_pred_svm = svm_model.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy:", accuracy_svm)

y_pred_lr = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy:", accuracy_lr)

y_pred_xgb = xgb_model.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print("XG Boost Accuracy:", accuracy_xgb)
```

It looks like the best models amongs chosen were XG Boost and Random Forest.
